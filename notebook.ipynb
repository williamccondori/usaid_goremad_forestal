```python
import geopandas
import pandas as pd
from sqlalchemy import create_engine, inspect
```

```python
def get_sql_query(tabla: str, esquema: str = 'public') -> str:
    sql = f"SELECT * FROM {esquema}.{tabla}"
    return sql
```
